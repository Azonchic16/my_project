```python
import numpy as np
import matplotlib.pyplot as plt


h1 = 10**(-2)
h2 = 10**(-3)
x0 = 2
eps = 1
a = 0
b = 10
T = 2

const = 0.7
v = -1
tau1 = h1*const/np.abs(v)
tau2 = h2*const/np.abs(v)

ksi = lambda x, x0, eps: np.abs(x - x0) / eps
phi1 = lambda x, x0, eps: np.heaviside(1 - ksi(x, x0, eps), 0)
phi2 = lambda x, x0, eps: phi1(x, x0, eps) * (1 - (ksi(x, x0, eps))**2)
phi3 = lambda x, x0, eps: phi1(x, x0, eps) * np.exp((-(ksi(x, x0, eps))**2) / np.abs(1 - (ksi(x, x0, eps))**2))
phi4 = lambda x, x0, eps: phi1(x, x0, eps) * (np.cos((np.pi/2) * ksi(x, x0, eps)))**3
real = lambda x, t, v, x0, eps: np.heaviside(x - v*t, 0) * phi4(x-v*t, x0, eps) + np.heaviside(v*t - x, 0) * phi4(t - (x / v), x0, eps)

x1 = np.linspace(a, b, int((b - a) / h1) + 1)
M1 = len(x1)
x2 = np.linspace(a, b, int((b - a) / h2) + 1)
M2 = len(x2)
t1 = np.linspace(0, T, int(T/tau1) + 1)
T1 = len(t1)
t2 = np.linspace(0, T, int(T/tau2) + 1)
T2 = len(t2)
u0x1 = phi4(x1, x0, eps)
u0x2 = phi4(x2, x0, eps)
mu1 = phi4(t1, x0, eps)
mu2 = phi4(t2, x0, eps)
b_ = a - h1


A1 = (v * tau1) / (2 * h1) + (tau1**2 * v**2) / (2 * h1**2)        # коэффициент при u_n,j-1
B1 = 1 - (2 * tau1**2 * v**2) / (2 * h1**2)               # коэффициент при u_n,j
C1 = (tau1**2 * v**2) / (2 * h1**2) - (v * tau1) / (2 * h1) # коэффициент при u_n,j+1

r1 = np.array([])
for n in range(1,T1):
    u1 = np.zeros_like(x1)
    u1[0] = mu1[n]
    u1[1:M1 - 1] = A1* u0x1[:-2] + B1 * u0x1[1:-1] + C1 * u0x1[2:]
    u1[-1] = u0x1[M1-1] - (v*tau1/h1) * (u0x1[M1-1] - u0x1[M1 - 2])
    if n == 1:
        plt.plot(x1, real(x1, t1[n], v, x0, eps), color='g', label='аналитическое')
        plt.plot(x1, u1, color='b', label='численное, h=0.01')
        plt.ylim(-0.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f't = {round(t1[n], 3)}')
        plt.grid()
        plt.legend()
        plt.show()
    elif n == T1 // 2:
        plt.plot(x1, real(x1, t1[n], v, x0, eps), color='g', label='аналитическое')
        plt.plot(x1, u1, color='b',label='численное, h=0.01')
        plt.ylim(-0.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f't = {round(t1[n], 3)}')
        plt.grid()
        plt.legend()
        plt.show()
    elif n == T1 - 1:
        plt.plot(x1, real(x1, t1[n], v, x0, eps), color='g', label='аналитическое')
        plt.plot(x1, u1, color='b', label='численное, h=0.01')
        plt.ylim(-0.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f't = {round(t1[n], 3)}')
        plt.grid()
        plt.legend()
        plt.show()
    r1 = np.append(r1, np.max(np.abs(real(x1, t1[n], v, x0, eps) - u1)))
    u0x1 = u1
    
A2 = (v * tau2) / (2 * h2) + (tau2**2 * v**2) / (2 * h2**2)
B2 = 1 - (2 * tau2**2 * v**2) / (2 * h2**2)
C2 = (tau2**2 * v**2) / (2 * h2**2) - (v * tau2) / (2 * h2)
```

```python
r2 = np.array([])
for n in range(1,T2):
    u2 = np.zeros_like(x2)
    u2[0] = mu2[n]
    u2[1:M2 - 1] = A2 * u0x2[:-2] + B2 * u0x2[1:-1] + C2 * u0x2[2:]
    u2[-1] = u0x2[M2-1] - (v*tau2/h2) * (u0x2[M2-1] - u0x2[M2 - 2])
    if n == 1:
        plt.plot(x2, real(x2, t2[n], v, x0, eps), color='g', label='аналитическое')
        plt.plot(x2, u2, color='b', label='численное, h=0.01')
        plt.ylim(-0.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f't = {round(t2[n], 3)}')
        plt.grid()
        plt.legend()
        plt.show()
    elif n == T2 // 2:
        plt.plot(x2, real(x2, t2[n], v, x0, eps), color='g', label='аналитическое')
        plt.plot(x2, u2, color='b', label='численное, h=0.01')
        plt.ylim(-0.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f't = {round(t2[n], 3)}')
        plt.grid()
        plt.legend()
        plt.show()
    elif n == T2 - 1:
        plt.plot(x2, real(x2, t2[n], v, x0, eps), color='g', label='аналитическое')
        plt.plot(x2, u2, color='b', label='численное, h=0.01')
        plt.ylim(-0.2, 1.2)
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f't = {round(t2[n], 3)}')
        plt.grid()
        plt.legend()
        plt.show()
    r2 = np.append(r2, np.max(np.abs(real(x2, t2[n], v, x0, eps) - u2)))
    u0x2 = u2
```

```python
plt.plot(t1[1:], r1, color='g', label='h=0.01')
plt.plot(t2[1:], r2, color='r', label='h=0.001')
plt.grid()
plt.ylabel('погрешность')
plt.xlabel('t')
plt.legend()
plt.show()
```
