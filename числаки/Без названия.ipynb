```python
class Fraction:

    def __init__(self, *p):
        if len(p) == 1:
            lst = (p[0]).split('/')
            self.ch = abs(int(lst[0]))
            self.zn = abs(int(lst[1]))
            num = int(lst[0])
            den = int(lst[1])
            self.__minus__(num, den)
        else:
            self.ch = abs(p[0])
            self.zn = abs(p[1])
            num = p[0]
            den = p[1]
            self.__minus__(num, den)
        self.__need__()

    def __need__(self):
        n = self.ch
        k = self.zn
        while n != 0 and k != 0:
            if n > k:
                n = n % k
            else:
                k = k % n
        nod = n + k
        if nod != 1:
            self.ch = self.ch // nod
            self.zn = self.zn // nod
            
    def __minus__(self, num, den):
        if (num < 0 and den < 0) or (num > 0 and den > 0):
            self.min = 1
        elif (num < 0 and den > 0) or (num > 0 and den < 0):
            self.min = -1
            
    def numerator(self, *p):
        if len(p) == 0:
            return self.ch
        else:
            if p[0] < 0:
                self.ch = abs(p[0])
                if self.min == -1:
                    self.min = 1
                else:
                    self.min = -1
                self.__need__()
            else:
                self.ch = p[0]
                self.__need__()
 
    def denominator(self, *p):
        if len(p) == 0:
            return self.zn
        else:
            if p[0] < 0:
                self.zn = abs(p[0])
                if self.min == -1:
                    self.min = 1
                else:
                    self.min = -1
                self.__need__()
            else:
                self.zn = p[0]
                self.__need__()
            
    def __str__(self):
        if self.min == -1:
            return f'-{self.ch}/{self.zn}'
        else:
            return f'{self.ch}/{self.zn}'
    
    def __repr__(self):
        if self.min == -1:
            return f"Fraction('-{self.ch}/{self.zn}')"
        else:
            return f"Fraction('{self.ch}/{self.zn}')"
    
    def __neg__(self):
        if self.min == -1:
            return Fraction(f'{self.ch}/{self.zn}')
        else:
            return Fraction(f'-{self.ch}/{self.zn}')
        
    def __add__(self, other):
        ch = self.min * self.ch * other.zn + other.min * other.ch * self.zn
        zn = self.zn * other.zn
        Fraction(ch, zn).__need__()
        return Fraction(ch, zn)
    
    def __sub__(self, other):
        ch = self.min * self.ch * other.zn - other.min * other.ch * self.zn
        zn = self.zn * other.zn
        Fraction(ch, zn).__need__()
        return Fraction(ch, zn)
    
    def __iadd__(self, other):
        self.ch = self.min * self.ch * other.zn + other.min * other.ch * self.zn
        self.zn = self.zn * other.zn
        self.__need__()
        self.__init__(self.ch, self.zn)
        return self
    
    def __isub__(self, other):
        self.ch = self.min * self.ch * other.zn - other.min * other.ch * self.zn
        self.zn = self.zn * other.zn
        self.__need__()
        self.__init__(self.ch, self.zn)
        return self
    
a = Fraction(1, 8)
c = b = Fraction(-3, 8)
b -= a
print(a, b, c, b is c)
```
