```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x0 = 0
b = 1
h = (10)**(-2)
a = 1 / np.sqrt(2)
a_l = 1
b_l = 0
a_r = 1
b_r = -1
c = 0.7
tau = c * h / a
x = np.linspace(x0, b, int((b - x0) / h) + 1)
T = 1
t1 = np.linspace(0, T, int(T / tau) + 1)

f = lambda x, t: 1 + (2 * x**2 - t**2) * (1 + 2 * (np.tan(x * t))**2) / (4 * np.cos(x * t))
phi = lambda x: x**2 + 1 / 2
phi_der_2 = lambda x: 2
psi = lambda x: -2 * x
mu_l = lambda t: t**2 + 1 / 2
mu_r = lambda t: t**2 - 1 + (1 - t * np.tan(t)) / (2 * np.cos(t))
real = lambda x, t: (x - t)**2 + 1 / (2 * np.cos(x * t))

def u0():
    return phi(x)

def u1_first():
    return phi(x) + tau * psi(x)

def u1_second():
    return phi(x) + tau * psi(x) + (((tau)**2) / 2) * (a**2 * phi_der_2(x) + f(x, 0))

def gran1(u, t):
    left = (mu_l(t) + (b_l * u[1]) / h) / (a_l + b_l / h)
    right = (mu_r(t) + (b_r * u[-2]) / h) / (a_r + b_r / h)
    return (left, right)
    
def gran2(u, t):
    left = (mu_l(t) - 4 * b_l * u[1] / (2 * h) + b_l * u[2] / (2 *h)) / (a_l + 3 * b_l / (2 * h))
    right = (mu_r(t) + 4 * b_r * u[-2] / (2 * h) - b_r * u[-3] / (2 *h)) / (a_r + 3 * b_r / (2 * h))
    return (left, right)

u0x = u0()
u1x = u1_first()          #   
def method(u0, u1, x, t):
    u = np.zeros_like(x)
    u[1:-1] = 2 * u1[1:-1] - u0[1:-1] + (tau * a / h) ** 2 * (u1[2:] - 2 * u1[1:-1] + u1[:-2]) + tau**2 * f(x[1:-1], t - tau)
    u[0] = gran1(u, t)[0]  #   
    u[-1] = gran1(u, t)[1] #   
    return u

fig, ax = plt.subplots(2, 1, figsize=(8, 10))
ax[0].grid(True)
ax[0].set_xlim(x0 - h, b + h)
ax[0].set_ylim(0, 2)
ax[0].set_xlabel('x')
ax[0].set_ylabel('u')
ax[1].set_label('')
ax[1].set_xlabel('t')
ax[1].set_ylabel('')
ax[1].set_xlim(0, 1)
ax[1].set_ylim(0, 0.005)
ax[1].grid(True)
(line1,) = ax[0].plot([], [], lw=3, color='b', label='')
(line2,) = ax[0].plot([], [], lw=3, color='g', label='')
(line3,) = ax[1].plot([], [], color='r')
ax[0].legend()

def init():
    line1.set_data([], [])
    line2.set_data([], [])
    line3.set_data([], [])
    return (line1, line2, line3, )

r = np.array([])
# r = np.append(r, np.max(np.abs(real(x, 0) - u0x)))
#r = np.append(r, np.max(np.abs(real(x, t1[1]) - u1x)))
def update(i):
    global u0x, u1x, r
    if i == 0:
        r = np.append(r, np.max(np.abs(real(x, i) - u0x)))
        line1.set_data(x, u0x)
        line2.set_data(x, real(x, i))
        line3.set_data(t1[:len(r)], r)
        return line1, line2, line3
    elif i == 1:
        r = np.append(r, np.max(np.abs(real(x, i) - u1x)))
        line1.set_data(x, u1x)
        line2.set_data(x, real(x, i))
        line3.set_data(t1[:len(r)], r)
        return line1, line2, line3
    else:
        u_res = method(u0x, u1x, x, i)
        r = np.append(r, np.max(np.abs(real(x, i) - u_res)))
        u0x = np.copy(u1x)
        u1x = np.copy(u_res)
        line1.set_data(x, u1x)
        line2.set_data(x, real(x, i))
        line3.set_data(t1[:len(r)], r)
        return line1, line2, line3

anim = FuncAnimation(
    fig, update, init_func=init, frames=t1, interval=1, blit=True, repeat=False
)

# ax[0].legend()
# ax[1].legend()
```
