```python
import numpy as np
import matplotlib.pyplot as plt

h1 = 10**(-2)
x0 = 2
eps = 1
a = 0
b = 10
T = 2
v = -4 # 4  -3
const = 0.7
tau1 = h1 * const / np.abs(v)

x1 = np.linspace(a, b, int((b - a) / h1) + 1)
M1 = len(x1)
t1 = np.linspace(0, T, int(T / tau1) + 1)
T1 = len(t1)

def ploting(x, t, u):
    plt.plot(x, real(x, t[n]), color='g', label='')
    plt.plot(x, u, color='b', label=f', h={h1}')
    plt.ylim(-0.2, 1.2)
    plt.xlabel('x')
    plt.ylabel('u')                              # 
    plt.title(f't = {round(t[n], 3)}')
    plt.grid()
    plt.legend()
    plt.show()

ksi = lambda x: np.abs(x - x0) / eps
phi1 = lambda x: np.heaviside(1 - ksi(x), 0)
phi2 = lambda x: phi1(x) * (1 - (ksi(x))**2)
phi3 = lambda x: phi1(x) * (np.cos((np.pi/2) * ksi(x)))**3

u_0 = phi1         #  

real = lambda x, t: u_0((x-v*t) % (b - a))

u0x1 = u_0(x1)

def f(u):        #  f
    return v * u

def f_u():      #   f  u
    return v

if f_u() > 0:
    def method(u0, u1):
        u1[-2] = u0x1[-2] - (v*tau1/h1) * (u0x1[-2] - u0x1[-3])
        u1[-1] =  u0x1[-1] - (v*tau1/h1) * (u0x1[-1] - u0x1[-2])
        u1[1] = u0x1[1] - (v*tau1/h1) * (u0x1[1] - u0x1[0])
        u1[0] = u1[-1]
else:
    def method(u0, u1):
        u1[1] = u0x1[1] - (v*tau1/h1) * (u0x1[2] - u0x1[1])
        u1[0] = u0x1[0] - (v*tau1/h1) * (u0x1[1] - u0x1[0])
        u1[-2] = u0x1[-2] - (v*tau1/h1) * (u0x1[-1] - u0x1[-2])
        u1[-1] = u1[0]

def a1():          # a_(i + 1/2)
    return f_u()

def a2():          # a_(i - 1/2)
    return f_u()

def u1_up(u): # u_(i + 1/2)
    res = np.heaviside(a1(), 0) * u[2:-2] + np.heaviside(-a1(), 0) * u[3:-1]
    res[np.isnan(res)] = 0
    return res

def u2_up(u): # u_(i - 1/2)
    res = np.heaviside(a2(), 0) * u[1:-3] + np.heaviside(-a2(), 0) * u[2:-2]
    res[np.isnan(res)] = 0
    return res

def F1_up(u):  # F_(i + 1/2)
    return f(u1_up(u))              
                                #   
def F2_up(u):  # F_(i - 1/2)
    return f(u2_up(u))

def u1_lax(u):
    res = (1 / 2) * (u[2:-2] + u[3:-1]) - (tau1 / (2 * h1)) * (f(u[3:-1]) - f(u[2:-2]))
    res[np.isnan(res)] = 0
    return res

def u2_lax(u):
    res = (1 / 2) * (u[2:-2] + u[1:-3]) - (tau1 / (2 * h1)) * (f(u[2:-2]) - f(u[1:-3]))
    res[np.isnan(res)] = 0
    return res

def F1_lax(u):
    return f(u1_lax(u))

def F2_lax(u):
    return f(u2_lax(u))

def slope1(u):
    if f_u() > 0:
        u_nan = (u[2:-2] - u[1:-3]) / (u[3:-1] - u[2:-2])
        u_nan[np.isnan(u_nan)] = 0
        return u_nan
    else:
        u_nan = (u[3:-1] - u[4:]) / (u[2:-2] - u[3:-1])
        u_nan[np.isnan(u_nan)] = 0
        return u_nan
    
def slope2(u):
    if f_u() > 0: 
        u_nan = (u[1:-3] - u[:-4]) / (u[2:-2] - u[1:-3])
        u_nan[np.isnan(u_nan)] = 0
        return u_nan
    else:
        u_nan = (u[2:-2] - u[3:-1]) / (u[1:-3] - u[2:-2])
        u_nan[np.isnan(u_nan)] = 0
        return u_nan
    
def limiter(r):
    res = (1.5 * (r + np.abs(r))) / (r + 2)
    where_nan = np.isnan(res)
    res[where_nan] = 0
    return res

for n in range(1,T1):
    u1 = np.zeros(M1)
    F1 = F1_up(u0x1) + limiter(slope1(u0x1)) * (F1_lax(u0x1) - F1_up(u0x1))
    F2 = F2_up(u0x1) + limiter(slope2(u0x1)) * (F2_lax(u0x1) - F2_up(u0x1))
    F = F1 - F2
    F[np.isnan(F)] = 0
    u1[2:-2] = u0x1[2:-2] - (tau1 / h1) * F
    method(u0x1, u1)
    if n == 1:
        ploting(x1, t1, u1)
    elif n == T1 // 2:
        ploting(x1, t1, u1)
    elif n == T1 - 1:
        ploting(x1, t1, u1)
    u0x1 = u1

    
```
