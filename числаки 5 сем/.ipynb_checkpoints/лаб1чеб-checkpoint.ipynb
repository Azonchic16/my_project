```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
def func1(a, b, n):
    x = np.linspace(a,b,n)
    f = lambda b: (np.log(b+2)*np.sin(b/2))/2 
    y = f(x)
    z, t = sp.symbols('z, t')
    polinom = 0
    for k in range(len(x)):
        p = 1
        for j in range(len(x)):
            if k != j:
                p = p*(t-x[j])/(x[k]-x[j])
        polinom = polinom + y[k]*p
    return sp.plot((polinom, (t, 0, 10)),(sp.log(z+2)*sp.sin(z/2)/2, (z,0,10)))
a = int(input())
b = int(input())
n = int(input())
    
print(func1(a,b,n+1))
```
