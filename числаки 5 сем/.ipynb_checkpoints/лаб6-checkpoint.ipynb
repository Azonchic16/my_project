```python
import numpy as np
import matplotlib.pyplot as plt
import math

f_real = lambda x: np.sin(x) + np.cos(x)

a = 0
b = 1
h = 0.05
x = np.arange(a, b+h, h)
def f1():
    u = [1.3818]
    ai = 1
    c = lambda x: 2 + np.cos(x)*h - np.sin(x)*(h**2)
    b = lambda x: 1 + np.cos(x)*h
    f = lambda x: -(h**2)*(1 - np.cos(x) - np.sin(x))
    ci = [c(x) for x in x]
    bi = [b(x) for x in x]
    fi = [f(x) for x in x]
    alpha = [1/(1+h)]
    beta = [0]
    for i in range(len(x)-1):
        alpha.insert(i+1, bi[i]/(ci[i] - alpha[i]*ai))
        beta.insert(i+1, (ai*beta[i] + fi[i])/(ci[i] - alpha[i]*ai))
    for i in range(-2, -len(x)-1, -1):
        u.insert(i, alpha[i+1]*u[i+1] + beta[i+1])
    return u

x1 = [(-h + i*h) for i in range(int((b-a)/h + 2))]
def f2():
    u = []
    u.append(1.3818)
    a = lambda x: 1 - math.cos(x)*h/2
    c = lambda x: -2 + math.sin(x)*(h**2)
    b = lambda x: 1 + math.cos(x)*h/2
    f = lambda x: (h**2)*(1 - math.cos(x) - math.sin(x))
    ai = [a(x1[i]) for i in range(len(x1))]
    ai[0] = 1
    ci = [c(x1[i]) for i in range(len(x1))]
    ci[0] = 2*h
    bi = [b(x1[i]) for i in range(len(x1))]
    bi[0] = -1
    fi = [f(x1[i]) for i in range(len(x1))]
    fi[0] = 0
    alpha = [1/(1+h)]
    beta = [0]
    for i in range(1,len(x1)):
        alpha.insert(i, -bi[i-1]/(ci[i-1] + alpha[i-1]*ai[i-1]))
        beta.insert(i, (fi[i-1] - ai[i-1]*beta[i-1])/(ci[i-1] + alpha[i-1]*ai[i-1]))
    for i in range(-2,-len(x1)-1,-1):
        u.insert(i, alpha[i+1]*u[i+1] + beta[i+1])
    return np.array(u)

u = f2()[1:]
print(len(u))
fig, axs = plt.subplots(figsize = (8, 8),nrows= 2, ncols= 2)
axs[0,0].plot(x, f1())
axs[0,0].plot(x, f_real(x))
axs[0,1].scatter(x, abs(np.array(f1()) - f_real(x)))
axs[1,0].plot(x1[1:], u)
axs[1,1].scatter(x1[1:], abs(u - f_real(x1[1:])))
axs[1,0].plot(x1[1:], f_real(x1[1:]))
print(max(abs(np.array(f1()) - f_real(x))))

```
