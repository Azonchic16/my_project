```python
import numpy as np
import matplotlib.pyplot as plt
import math
a = int(input("левый конец отрезка: "))
b = int(input("правый конец отрезка: "))
n = int(input("на сколько отрезков разбить отрезок: "))
x = np.linspace(a,b,n+1)
f = lambda b: np.log(2+x*(1+np.cos(x)))
y = f(x)
def func1(t: float):
    polinom = 0
    for k in range(len(x)):
        p = 1
        for j in range(len(x)):
            if k != j:
                p = p*(t-x[j])/(x[k]-x[j])
        polinom = polinom + y[k]*p
    return polinom

    
c = np.linspace(0,10,100)
g = [func1(i) for i in c]
x_err = np.array([x[i] + (b-a)/(4*n) for i in range(0, len(x)-1)])
pol_chetv = np.array([func1(elem) for elem in x_err])
y_chetv = f(x_err)
r = abs(pol_chetv - y_chetv)
fig, axs = plt.subplots(nrows= 2 , ncols= 2 )
axs[0, 0].plot(c,g,label = 'полином')
axs[0, 0].plot(c,f(c),label = 'функция')
axs[0, 0].legend(fontsize = 14)
axs[0, 0].errorbar(x_err, pol_chetv, yerr=r, linestyle='none', ecolor = 'k',elinewidth=0.5, capsize=2, capthick=1)
axs[0, 0].grid(which='major')
axs[1, 0].scatter(x_err, r)
axs[1, 0].grid(True)

q = [0.5*(a+b)+0.5*(b-a)*np.cos(np.pi*(2*k+1)/(2*n+2)) for k in range(n+1)]
x_cheb = np.array(q[::-1])
y_cheb = f(x_cheb)

def func2(t: float):
    polinom = 0
    for k in range(len(x_cheb)):
        p = 1
        for j in range(len(x_cheb)):
            if k != j:
                p = p*(t-x_cheb[j])/(x_cheb[k]-x_cheb[j])
        polinom = polinom + y_cheb[k]*p
    return polinom

    
g_cheb = [func2(i) for i in c]
x_err_cheb = np.array([x_cheb[k] + (x_cheb[k+1] - x_cheb[k])/4 for k in range(n)])
pol_chetv_cheb = np.array([func2(elem) for elem in x_err_cheb])
y_chetv_cheb = f(x_err_cheb)
r_cheb = abs(y_chetv_cheb - pol_chetv_cheb)
axs[0,1].plot(c,g_cheb,label = 'полином чебышева')
axs[0,1].plot(c,f(c),label = 'функция')
axs[0,1].legend(fontsize = 14)
axs[0,1].grid(True)
axs[0,1].errorbar(x_err_cheb, pol_chetv_cheb, yerr=r_cheb, linestyle='none', ecolor = 'k',elinewidth=0.5, capsize=2, capthick=1)
axs[1,1].scatter(x_err_cheb, r_cheb)
axs[1,1].grid(True)
fig.set_size_inches(14,7)
plt.show()
print(x_cheb)
```
